````python
#We don't need Event_gender and Country_code for analysing this dataset so drop these columns.
df= df.drop('Event_gender', axis = 1)
df= df.drop('Country_Code', axis = 1)
````

````python
#Change the datatype of year 'int'
df['Year'] = df['Year'].astype(int)
````

# Exploratory Data Analysis (EDA)

````python
# Which city hosted maximum number of olympics
Unique_year = df[["City", 'Year']]
Unique_year = Unique_year.drop_duplicates('Year')
````

````python
Unique_year
````

````python
# Which city hosted most events.
City = df['City'].value_counts()
City.columns = ['City', 'Count']
````

````python
plt.figure(figsize=(10,5))
plt.bar(City.index, City.values)
plt.xticks(rotation=45)
plt.title("City hosted maximum number of events")
plt.xlabel("City")
plt.show()
````

````python
# Understand the events themselves.
q3_data = df[['Sport', 'Discipline', 'Event']].drop_duplicates()
print("Total number of unique events are held so far are :",len(q3_data))
q3_data
````

````python
q3_data = q3_data.groupby(['Sport'])['Sport'].size()
plt.figure(figsize = (10,4))
q3_data.plot.bar(x = 'Sport', y = 'Count')
````

````python
# Which Athlete has win most medal from given period?
athlete_medal_count = df.groupby('Athlete')['Medal'].count().sort_values(ascending=False)

# Plotting the top 10 athletes with most medals
plt.figure(figsize=(10, 6))
athlete_medal_count.head(10).plot(kind='bar', color='silver')
plt.title("Top 10 Athletes by Medal Count")
plt.xlabel("Athlete")
plt.ylabel("Total Medals")
plt.show()
````

````python
# Gender ratio in winning teams?
q4_data = df.groupby(['Gender'])['Gender'].count()
plt.figure(figsize = (12,2))
q4_data.plot.barh(x = 'Athlete', y = 'Count')
````

````python
q4_data = df[['Event', 'Gender']]
q4_data = q4_data.groupby(['Event', 'Gender'])['Gender'].count()
print(q4_data.head(20))
````

# Data Visualization

````python
# Plotting the top 10 countries by medals
medals_by_country = df.groupby('Country')['Medal'].count().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
medals_by_country.head(10).plot(kind='bar', color='gold')
plt.title("Top 10 Countries by Medal Count")
plt.xlabel("Country")
plt.ylabel("Total Medals")
plt.show()
````

````python
# Plotting the trend of medals won over the years
medals_over_years = df.groupby('Year')['Medal'].count()
plt.figure(figsize=(10, 6))
plt.plot(medals_over_years.index, medals_over_years.values, marker='o', linestyle='-', color='b')
plt.title("Total Medals Won Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Medals")
plt.grid(True)
plt.show()
````

````python
# Plotting gender distribution
gender_distribution = df['Gender'].value_counts()
plt.figure(figsize=(6, 4))
gender_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], explode=[0.05, 0])
plt.title("Gender Distribution in Olympics Events")
plt.ylabel('')
plt.show()
````

````python
# Medal won by category by country wise
temp1 = df.groupby(['Country', 'Medal'])['Medal'].count().reset_index(name = 'Count').sort_values(ascending = [True, False],by = ['Country','Count'])
````

````python
print(temp1.head(20))
````

````python
# Plotting Medal distribution
Medal_distribution = df['Medal'].value_counts()
plt.figure(figsize=(6, 4))
Medal_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff', '#c2c2f0'], explode=[0, 0, 0])
plt.title("Medal Distribution in Olympics Events")
plt.ylabel('')
plt.show()
````

# Conclusion:
## Insights from Olympic Data Analysis (1976-2008)

The analysis of the Summer Olympic Games data from 1976 to 2008 has yielded several key insights into the trends and performance patterns of participating nations and athletes. We observed a consistent increase in the total number of medals awarded over the years, suggesting an expansion of events or participant numbers within the Olympics.

Geographically, the United States stands out as a dominant force, consistently leading in overall medal counts across multiple sports. Other nations like the Soviet Union (and later Russia) and Australia also demonstrated strong performances. The analysis further highlighted specific national strengths, such as Cuba's dominance in Boxing and Korea, South's prowess in Archery and Taekwondo, as well as China's strong performance in Table Tennis and Weightlifting.

From an individual athlete's perspective, the data revealed exceptional multi-medallists like Michael Phelps, Birgit Fischer, and Nikolay Andrianov, underscoring their remarkable athletic achievements and sustained excellence over several Olympic cycles.

Gender distribution in medal wins showed a significant participation and success rate for both men and women, reflecting the evolving inclusivity of the Olympic Games. The data also uncovered instances where athletes successfully competed and won medals across different sports or disciplines, showcasing their versatility and adaptability.

Overall, this analysis provides a valuable historical overview of Olympic performance, identifying trends in medal distribution, national sporting strengths, and the achievements of individual athletes. These insights can be instrumental for understanding the historical landscape of Olympic competition and for informing future strategic planning in sports development.
